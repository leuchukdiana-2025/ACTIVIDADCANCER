```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv('data.csv')
df.info()
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].map({'B':0, 'M':1})
```

```python
x = df[['radius_mean', 'texture_mean', 'smoothness_mean']]
print (x)

y = df['diagnosis']
scaler = StandardScaler() #Escalo el modelo
XScaled = scaler.fit_transform(x)

Kn=KNeighborsClassifier(n_neighbors=5) #Creacion de modelo
Kn.fit(XScaled, y )
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
input = [[ 20.57, 17.77, 0.08474]] #DOY VALORES PARA PREDECIR
input_df = pd.DataFrame(input, columns=['radius_mean', 'texture_mean', 'smoothness_mean'])
inputScaler = scaler.transform(input_df)
output= knn.predict(inputScaler)
if output == 0:
    print("Predicción manual: Tumor BENIGNO")
else:
    print("Predicción manual: Tumor MALIGNO")
```
